# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# Plot the histogram
QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

# Plot the histogram
plt.hist(df['ConvertedComp'], bins=30, color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Converted Annual Compensation (USD)')
plt.ylabel('Number of Respondents')
plt.title('Distribution of Converted Annual Compensation')

plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
# Create a box plot
QUERY = """
SELECT Age
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
plt.boxplot(df['Age'])

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Distribution')
plt.title('Box Plot of Age')
# Create the boxplot
sns.boxplot(x=df['Age'])

# Add labels and title
plt.xlabel('Age')
plt.title('Boxplot of Age')

plt.show()

plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Create a scatter plot
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
plt.scatter(df['Age'], df['WorkWeekHrs'])

# Add labels and title
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')


plt.show()


```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Scatter plot with bubble size based on Age
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*20, alpha=0.7)  # Adjust factor for bubble size

# Add labels and title
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot: WorkWeekHrs vs. CodeRevHrs (Bubble size = Age)')

# Optional: Add colormap for additional dimension (consider using seaborn for colormaps)
# plt.cm.coolwarm(df['Age'])  # Example colormap, adjust as needed

plt.grid(True)  # Add grid lines for better readability
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Get the top 5 databases
QUERY = """
SELECT DatabaseDesireNextYear, Respondent, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY,conn)
top_5 = df.nlargest(5, 'Respondent')

# Create the pie chart
plt.pie(top_5['Respondent'], labels=top_5['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Top 5 Databases to Learn Next Year')

# Display the chart
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Filter for age group 30 to 35
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age >= 30 AND Age <= 35
"""
df = pd.read_sql_query(QUERY,conn)
df_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median WorkWeekHrs and CodeRevHrs for the age group
median_values = df_filtered[['WorkWeekHrs', 'CodeRevHrs']].median().to_frame().T

# Create the stacked bar chart
median_values.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')


plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Filter for age group 45 to 60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age >= 45 AND Age <= 60
"""
df = pd.read_sql_query(QUERY,conn)

# Calculate the median ConvertedComp for each age group
df_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Group by Age and calculate median ConvertedComp
median_comp = df_filtered.groupby('Age')['ConvertedComp'].median()

# Create the line plot
median_comp.plot(kind='line', marker='o', linestyle='-', figsize=(10, 6))

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')

plt.show()


```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Create horizontal bar chart
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY,conn)

# Replace 'Country' with 'MainBranch' for the y-axis
plt.barh(df['MainBranch'], df['Count'])

# Add labels and title
plt.xlabel('Count')  # Changed label to reflect count
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branch') # Adjusted title

# Show chart with gridlines
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()

```
